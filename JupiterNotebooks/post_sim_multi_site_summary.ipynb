```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

canada_df= pd.read_csv("/home/mremington/Documents/therma_sim/therma_sim/src/therma_sim/Results/Canada_Model.csv")
nebraska_df= pd.read_csv("/home/mremington/Documents/therma_sim/therma_sim/src/therma_sim/Results/Nebraska_Model.csv")
texas_df= pd.read_csv("/home/mremington/Documents/therma_sim/therma_sim/src/therma_sim/Results/Texas_Model.csv")
```

```python
# Create the figure and axes
fig, axes = plt.subplots(2, 2, figsize=(12, 6))
suptitle = "Population Dynamics Comparison: Canada, Nebraska, Texas\nOver 6 Year Simulation Period"
fig.suptitle(suptitle, fontsize=16)

# Define datasets and colors
datasets = {
    "Canada": (canada_df, 'green'),
    "Nebraska": (nebraska_df, 'orange'),
    "Texas": (texas_df, 'red')
}

# Plot: Rattlesnakes Over Time
for label, (df, color) in datasets.items():
    axes[0, 0].plot(df["Time_Step"], df["Rattlesnakes"], label=f"{label}", color=color)
axes[0, 0].set_title("Rattlesnakes Over Time Step")
axes[0, 0].set_xlabel("Time Step")
axes[0, 0].set_ylabel("Count")
axes[0, 0].legend()

# Plot: Krats Over Time
for label, (df, color) in datasets.items():
    axes[0, 1].plot(df["Time_Step"], df["Krats"], label=f"{label}", color=color)
axes[0, 1].set_title("Krats Over Time Step")
axes[0, 1].set_xlabel("Time Step")
axes[0, 1].set_ylabel("Count")


# Plot: Rattlesnake Density
for label, (df, color) in datasets.items():
    axes[1, 0].plot(df["Time_Step"], df["Rattlesnakes_Density"], label=f"{label}", color=color)
axes[1, 0].set_title("Density of Rattlesnakes Over Time Step")
axes[1, 0].set_xlabel("Time Step")
axes[1, 0].set_ylabel("Density")


# Plot: Krats Density
for label, (df, color) in datasets.items():
    axes[1, 1].plot(df["Time_Step"], df["Krats_Density"], label=f"{label}", color=color)
axes[1, 1].set_title("Density of Krats Over Time Step")
axes[1, 1].set_xlabel("Time Step")
axes[1, 1].set_ylabel("Density")
axes[1, 1].set_ylim(0, 14)
axes[1, 1].legend()

# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Space for suptitle
plt.show()
```
